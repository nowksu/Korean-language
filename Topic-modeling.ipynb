# Preprocessing & Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
```

```python
dtm = cv.fit_transform(npr['Article'])
```

```python
dtm
```

# Implementing LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
```

We want to have 7 topic groups by using 'n_components'.

```python
LDA = LatentDirichletAllocation(n_components=7,n_jobs=-1, verbose=3, random_state=42)
```

```python
LDA.fit(dtm)
```

# Grab the Topics

```python
len(LDA.components_)
```

```python
LDA.components_.shape
```

This confirms that we have 7 topic groups and there are 54,777 unique words. Also, each element
represents how frequent a word is for a topic group. The higher the value is, the more frequent the
corresponding word is.


```python
LDA.components_
```

```python
len(LDA.components_[0])
```

We grap important scores for the first topic.

```python
single_topic = LDA.components_[0]
```

```python
# argsort() --> index positions sorted from least to greatest
# Returns the indices that would sort this array.
single_topic.argsort()
```

```python
# Word least representative of this topic
single_topic[2475]
```

```python
# Word most representative of this topic
single_topic[42993]
```

# Grab most frequent words per topic


```python
# Top 10 words for this topic:
# Last 10 values of argsort
single_topic.argsort()[-10:]
```

```python
top_ten_words = single_topic.argsort()[-10:]
```

```python
for index in top_ten_words:
 print(cv.get_feature_names()[index])
```

Let's view all the 7 topics found.


```python
for index,topic in enumerate(LDA.components_):
 print("THE TOP 15 WORDS FOR TOPIC #{}".format(index))
 print([cv.get_feature_names()[i] for i in topic.argsort()[-15:]])
 print('\n')
```

We began with 7 topics. It's up to us to decide the optimal number of topics in LDA. If we think
that a certain topic seems to have a mix of multiple topics, we can increase the number of topics.
We can reduce the number of topics if we think certain topics are overlaped. Some use statistics
such as coherence score to find the optimal number of topics. Also, recent topic models try to
incorporate the optimal number of topics in the estimation process.

The topic modeling analysis does not tell us what each topic represents. It's up to us to label
each topic. We will use top keywords to intepret each topic group
